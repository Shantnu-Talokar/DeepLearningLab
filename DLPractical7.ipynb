```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

breast_cancer = load_breast_cancer()
X = breast_cancer.data
y = breast_cancer.target

df = pd.DataFrame(X, columns=breast_cancer.feature_names)

display(df.head())
display(df.shape)

from sklearn.decomposition import PCA
pca = PCA()
pca.fit(df)
df_pca = pca.transform(df)

display(df_pca.shape)
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_explained_variance = explained_variance_ratio.cumsum()

print("Explained Variance Ratio for each component:")
print(explained_variance_ratio)

print("\nCumulative Explained Variance:")
print(cumulative_explained_variance)
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=y, cmap='viridis', alpha=0.6)
plt.title('PCA Transformed Data (First 2 Components)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Target Variable')

plt.subplot(1, 2, 2)
plt.scatter(df['mean radius'], df['mean texture'], c=y, cmap='viridis', alpha=0.6)
plt.title('Original Data (Mean Radius vs Mean Texture)')
plt.xlabel('Mean Radius')
plt.ylabel('Mean Texture')
plt.colorbar(label='Target Variable')

plt.tight_layout()
plt.show()
```
